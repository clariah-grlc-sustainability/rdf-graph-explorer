```python
from rdflib import Graph, plugin
from rdflib.store import Store
from rdflib.plugins.stores.sparqlstore import SPARQLStore

# Set the SPARQL endpoint URL
# sparql_endpoint = "https://publications.europa.eu/webapi/rdf/sparql"
sparql_endpoint = "https://dbpedia.org/sparql"

# Create a SPARQLStore and associate it with a Graph
store = SPARQLStore(sparql_endpoint)
g = Graph(store=store)

# Define the SPARQL query string
list_of_classes_query = """
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
SELECT DISTINCT ?x
WHERE {
    ?class a rdfs:Class .
    ?x a ?class .
}
LIMIT 500
"""

leaf_classes_query = """
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>

SELECT DISTINCT ?x
WHERE {
    ?x a rdfs:Class .
    FILTER NOT EXISTS {
        ?subClass rdfs:subClassOf ?x .
    }
}
LIMIT 100 OFFSET 40
"""

number_of_classes_query = """
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>

SELECT (COUNT(DISTINCT ?class) AS ?x)
WHERE {
    ?class a rdfs:Class .
}
"""

# class_entity = "<http://dbpedia.org/ontology/Actor>"

# metadata_properties_query = """
# PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
# PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>

# SELECT DISTINCT ?x
# WHERE {
#     ?x a rdf:Property ;
#             rdfs:domain {} ;
#             rdfs:range rdfs:Literal .
# }
# """.format(class_entity)

test_query= """
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>

SELECT DISTINCT ?x
WHERE {
    ?x a rdf:Property ;
            rdfs:domain <http://dbpedia.org/ontology/Person> .
}
"""

only_data_properties_query = """
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>

SELECT DISTINCT ?x
WHERE {
  ?subject ?x ?object .
  FILTER (isLiteral(?object))
}
LIMIT 300
"""

# Perform the query against the remote SPARQL endpoint
results = g.query(only_data_properties_query)

# Print the results row by row
idx = 1
for row in results:
    print(str(idx)+'.',  f"{row['x']}")
    idx += 1
```
